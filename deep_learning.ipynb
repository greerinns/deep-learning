## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

Targets vs Features:

*   IS_SUCCESSFUL - Target
*   Remaining Columns - Features



```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = application_df.drop(columns=['EIN','NAME'])
application_df.head(10)
```

```python
# Determine the number of unique values in each column.
application_df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts for binning
app_counts = application_df.APPLICATION_TYPE.value_counts(sort=False)
#sort = False to keep the unique types in the same order
app_counts
```

```python
unique_apps = application_df.APPLICATION_TYPE.unique()
# grabbing a list of the application type values for those counts
unique_apps
```

```python
# Choose a cutoff value of 200 and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
[application_types_to_replace = []
for i in range(len(unique_apps)):]
  if app_counts[i] <200:
    application_types_to_replace.append(unique_apps[i])
  else:
    pass
# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
class_counts = application_df.CLASSIFICATION.value_counts(sort=False)
#sort = False to keep the unique types in the same order
class_counts
```

```python
unique_class = application_df.CLASSIFICATION.unique()
# grabbing a list of the application type values for those counts
unique_class
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
class_counts_over_1 = [count for count in class_counts if count>1]
class_counts_over_1
```

```python
# Choose a cutoff value of 800 and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
classifications_to_replace = []
for i in range(len(unique_class)):
  if class_counts[i] <800:
    classifications_to_replace.append(unique_class[i])
  else:
    pass
# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Look at ASK AMT value counts to see if there is a use for binning
ask_counts = application_df.ASK_AMT.value_counts(sort=True)
ask_counts
# no need for binning, not many of each besides 5000
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
dummied_df = pd.get_dummies(application_df)
categorical_cols = dummied_df.columns
# looking at the new column names
categorical_cols
```

```python
dummied_df
```

```python
# Split our preprocessed data into our features and target arrays
X = dummied_df.drop(columns=['IS_SUCCESSFUL'])
y = dummied_df['IS_SUCCESSFUL']
# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=78)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=60, activation="relu", input_dim=43))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=15, activation="relu"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy",optimizer="adam",metrics=["accuracy"])
```

```python
# Train the model
trained_model = nn.fit(X_train_scaled, y_train, epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Export our model to HDF5 file
nn.save('AlphabetSoupCharity.h5')
```
